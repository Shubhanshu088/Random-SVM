```python
import pandas as pd
import matplotlib.pyplot
import numpy as np
```

```python
air=pd.read_csv(r'D:\Data_Analysis\Machine Learning\assign\air.csv')
```

```python
## let's first find in which europen country have more exited customer
country= air.groupby('Geography')['Exited'].count()
country
```

```python
## No null values let's convert object into dummy variable
```

```python
air=pd.get_dummies(air,columns=['Geography'])
air=pd.get_dummies(air,columns=['Gender'],drop_first=True)
```

```python
air.head()
```

```python
air.corr()
```

```python
## before making any model we have to make dependent and independent variable
```

```python
y=air[['Exited']]
x=air.drop(columns=['Exited'])
```

```python
## now split the data 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.85,random_state=4321)
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(n_estimators=1000,random_state=1234)
rf=model.fit(x_train,y_train)
```

```python
y_test['predicted']=model.predict(x_test)
y_test.head()
```

```python
from sklearn.metrics import accuracy_score
print('accuracy score of random forest model',accuracy_score(y_test['Exited'],y_test['predicted']))
```

```python
Y=air[['Exited']]
X=air.drop(columns=['Exited'])
x.head()
```
